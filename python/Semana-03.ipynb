```
pi = 3.14
x = 3.    # Se la indica que es un numero real con el . 
z = 3     # Este será un número entero
print(pi)  # Esto es un decimal
print(x)   # Es otro decimal
print(z)   # Es un entero

print("Un ejemplo de valor decimal es " + str(pi))
print("Un ejemplo de valor decimal es {}".format(pi))
```

# Variables

Una variable es el nombre que se refiere a un valor, la sentencia de asignación crea nuevas variables y les da valores. En Python a diferencia de otros lenguajes de programación no se declaran el tipo de variable.

## Reglas y convenciones
- No use simbolos especiales com @, # , ! , $, etc
- El primer caracter de la variable no puede ser un digito númerico.
- Debe empezar con una letra (a-z, A-Z) o un guión bajo (_)
- Los otros caracteres pueden ser letras, numeros o guión bajo
- Distingue mayúsculas de minúsculas

- No puede usar palabras reservadas

# Operadores

## Operadores Artiméticos

*   **+** : Sumar dos variables $a+b$
*   **-**  : Restar dos variables $a-b$
*   **\*** : Multiplicar dos variables $a\cdot b$
*   **/** : Dividir una variable por la otra $\frac{a}{b}$
*   **\*\*** : Una variable elevado a la otra $a^{b}$
*   **=** : Asignarle un valor a una variable $a \leftarrow b$
*   **%** : Obtiene el resto de una división $a \% b$

```
x = 3
y = 2
suma = x + y
resta = x - y
multi = x * y
div = x / y
potencia = x**y #pontencia al cuadrado
print(suma)
print(resta)
print(multi)
print(div)
print(potencia)
```

### Operador módulo
El operador módulo se utiliza con el símbolo **%**, y permite calcular el resto de un numero en la división por otro. Por ejemplo:

*   $5$ % $3$ $= 2$
*   $17$ % $10$ $= 7$
*   $9$ % $6$ $= 3$
*   $15$ % $5$ $= 0$

Podríamos por ejemplo preguntar si el resto de un numero **x** en la division por **y** es igual a 0 y así saber si **x** es multiplo de **y**:

```
x = 27
y = 9
print( x % y == 0 )
```


Si queremos saber el resto de la división entera, usamos el operador **módulo %** 

En muchos lenguajes de programación se indica con:  $ x\%y $

```
x = 5
y = 2
print(x // y) # Redondea division para abajo
print(x % y)  # Operación Módulo

#Se pueden mostrar varias cosas juntas si se separan con coma dentro de print()
print(x // y, "*", y, "+", x%y,"=", x)
```

## Operadores de Asignación
Python nos permite ahorrar código combinando el operando **=** con alguno de los anteriormente vistos:


*   ***=**  : Multiplica la variable $a \leftarrow a\cdot b$ 
*   **/=** : Divide la variable  $a \leftarrow \frac{a}{b}$ 
*   **+=** : Suma a la varaible  $a \leftarrow a+ b$ 
*   **-=** : Resta a la variable $a \leftarrow a- b$

```
x = 17
x += 3
x /= 10
print(x)
```

## Operadores de Comparación / Operadores relacionales
Existen operaciones donde se comparan las propiedades de una variable.
*   **<** : Menor que $a < b$
*   **>** : Mayor que $a > b$
*   **<=** : Menor o igual que $a \leq b$
*   **>=** : Mayor o igual que $a \geq b$
*   **==** : Es igual $a = b$
*   **!=** : Es distinto $a \neq b$

Al comprarar dos números se obtiene True o False (Verdadero o Falso) según si se cumple la condición:

```
x = 4
print(x >= 4)
print(x < 3)
```

```
x = int( input() )
x = 2*x
print( x%2 == 0 )
```

## Operadores Lógico

*   **and** : ¿se cumple a y b? => $a$ $and$ $b$
*   **or** : ¿se cumple a o b? => $a$ $or$ $b$
*   **not** : invertir el valor de a =>  $ not$ $a$




```
x = 4
y = x > 5 and x < 7
z = x > 5 or x < 7
k = not x > 5
print(y)
print(z)
print(k)
```

Estas condiciones pueden agruparse con ( condicion1 ) **and/or** ( condicion2 ), esto permite chequar condiciones complejas:

```
x = 10
y = 5
z = (y > x and 15 > x) or (y < x and 15 > x)
print(z)
```
