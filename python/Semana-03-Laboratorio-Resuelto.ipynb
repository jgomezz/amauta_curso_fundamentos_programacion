## Ejercicio 02

Utilizando operadores de asignación realice lo siguiente: 
- Reste a x el valor de y 
- Halle la división entera de x con respecto a y.
- Halle x elevado a la y 
- Halle el residuo de x entre y

Considere como valores iniciales:
- x = 5
- y = 3

Ahora utilice operadores aritméticos para hallar los mismos resultados. 

```
# Operadores de asignación
print("Operadores de asignación")
x = 5
y = 3

x -= y
print("x = {}, y = {}".format(x,y))

x **= y
print("x = {}, y = {}".format(x,y))

x %=y
print("x = {}, y = {}".format(x,y))

# Operadores artiméticos
print("Operadores artiméticos")

x = 5
y = 3

x = x - y
print("x = {}, y = {}".format(x,y))

x = x**y
print("x = {}, y = {}".format(x,y))

x = x%y
print("x = {}, y = {}".format(x,y))


```

## Ejercicio 03

Un grupo de bacterias se reproducen en un 20% con respecto al día anterior, si el día 1 hay 500 bacterias, ¿Cuantas habrá el día 5? Utilice operadores de asignación y las siguientes variables. 

- i = 0
- texto = "día"
- inicio = 500
- incremento = 0.2

```
i = 0
texto = "día"
inicio = 500
incremento = 0.2

# Día 1
siguiente_dia = inicio 
i+=1
print("En el {} {} hay {} bacterias".format(texto, i, siguiente_dia))

# Día 2
siguiente_dia *= (1+incremento)         
i+=1
print("En el {} {} hay {} bacterias".format(texto, i, siguiente_dia))

# Día 3
siguiente_dia *= (1+incremento)         
i+=1
print("En el {} {} hay {} bacterias".format(texto, i, siguiente_dia))

# Día 4
siguiente_dia *= (1+incremento)       
i+=1
print("En el {} {} hay {} bacterias".format(texto, i, siguiente_dia))

# Día 5
siguiente_dia *= (1+incremento)      
i+=1
print("En el {} {} hay {} bacterias".format(texto, i, siguiente_dia))
```
