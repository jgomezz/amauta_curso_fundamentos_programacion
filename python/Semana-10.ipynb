#### 3.2.- Funciones que requieren datos de ingreso y no retornan ningun valor
Este tipo de funciones requieren que se les pase valores, eso se realiza a través de un conjunto de variables conocidos como parámetros o argumentos.


```python
# Función para imprimir un mensaje personalizado, 
# se pasa a traves de un parámetro o argumento denominado
# mi_mensaje el valor que se desea imprimir.

def imprimir_mensaje(mi_mensaje):
  print(mi_mensaje)

imprimir_mensaje("Hola Mundo")
imprimir_mensaje("Ya estamos en Diciembre")
```

```python
# Ejercicio 04 : crear un función que me permita pasar como parámetro un valor booleano
#             si es True que imprima "Hola Mundo..!", si el False que imprima "Hello World...!"

# Definir la función
def imprimir_mensaje(var):
  if var :
    print("Hola Mundo..!")
  else:
    print("Hello World...!")

# Usar la función
imprimir_mensaje(True)
imprimir_mensaje(False)
```

```python
# Función que recibe mas de un parámetro de entrada

def sumar(a, b):
  print(a+b)

# Suma 2 numeros y lo muestra por pantalla
sumar(3, 4)  
```

#### 3.3.- Funciones que requieren datos de ingreso y retornan un valor
Este tipo de funciones requieren que se les pase valores, eso se realiza a través de un conjunto de variables conocidos como parámetros o argumentos, y retornan un valor o conjunto de valores al proceso que llamo a la función

```python
# Función que recibe 2 valores y devuelve la suma

def sumar(var1, var2):
  s = var1 + var2
  return s

ret = sumar(4, 50)
print("El valor de la suma es:", ret)
```

## 4.- Valores por defecto en una función
Algunas veces no necesitamos enviar todos los valores que una función necesita, cuando ese es el caso Python nos permite definir valores por defecto.

```python
# Función que increma un valor 
def incrementar(var1, inc = 1):
   tmp = var1 + inc
   return tmp

nro = 4
# Incrementa el valo en 1
print(incrementar(nro))

# Incrementa el valo en 10
print(incrementar(nro, 10))
```
