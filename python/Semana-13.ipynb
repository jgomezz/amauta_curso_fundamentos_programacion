```python
# Escritura de archivo

mi_archivo = open("mis_datos_nuevos.txt","w")

mes = 12     # Mes Diciembre
pi = 3.1415 # Constante PI

# Forma 1
str_forma_1 =  "Mes={},PI={}\n".format(mes,pi)
mi_archivo.write(str_forma_1)

# Forma 1
str_forma_2 =  "Mes=" + str(mes) + ",PI=" + str(pi) + "\n"
mi_archivo.write(str_forma_2)

mi_archivo.close()
```

## 6.- Atributos del objeto file

<center>
  <img width="900" src="https://raw.githubusercontent.com/jgomezm-tecsup/curso-introduccion-programacion/master/_assets/semana_13_atributos_objeto_file.png" />

```python
# Lectura de archivo, estados

mi_archivo = open("mis_datos.txt","r")

print("------------- ANTES ----------------")
print("Archivo  cerrado? = " , mi_archivo.closed)
print("Archivo  modo = " , mi_archivo.mode)
print("Archivo  nombre = " , mi_archivo.name)

mi_archivo.close()

print("------------- DESPUÊS ----------------")
print("Archivo  cerrado? = " , mi_archivo.closed)
print("Archivo  modo = " , mi_archivo.mode)
print("Archivo  nombre = " , mi_archivo.name)
```

```python
# Escritura de archivo, estados

mi_archivo = open("mis_datos.txt","w")

print("------------- ANTES ----------------")
print("Archivo  cerrado? = " , mi_archivo.closed)
print("Archivo  modo = " , mi_archivo.mode)
print("Archivo  nombre = " , mi_archivo.name)

mi_archivo.close()

print("------------- DESPUÊS ----------------")
print("Archivo  cerrado? = " , mi_archivo.closed)
print("Archivo  modo = " , mi_archivo.mode)
print("Archivo  nombre = " , mi_archivo.name)
```

```python
# Agregar datos a archivo, estados

mi_archivo = open("mis_datos.txt","a")

print("------------- ANTES ----------------")
print("Archivo  cerrado? = " , mi_archivo.closed)
print("Archivo  modo = " , mi_archivo.mode)
print("Archivo  nombre = " , mi_archivo.name)

mi_archivo.close()

print("------------- DESPUÊS ----------------")
print("Archivo  cerrado? = " , mi_archivo.closed)
print("Archivo  modo = " , mi_archivo.mode)
print("Archivo  nombre = " , mi_archivo.name)
```
