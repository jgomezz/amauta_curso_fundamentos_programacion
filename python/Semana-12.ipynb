```python

# Lectura de archivo, linea por linea

mi_archivo = open("datos.txt","r")

lineas = mi_archivo.readlines()

mi_archivo.close()

for linea in lineas:
  print(int(linea)) # convertimos a entero por el salto de linea

```

```python
# Lectura de archivo, sumando todos los valores de los archivos

mi_archivo = open("datos.txt","r")

lineas = mi_archivo.readlines()

mi_archivo.close()

suma = 0
for linea in lineas:
  suma = suma + int(linea) 
  
print(suma)
```

```python
# Lectura de archivo, linea por linea

mi_archivo = open("mensajes.txt","r")

msg1 = mi_archivo.readline()
print(msg1) 

msg2 = mi_archivo.readline()
print(msg2) 

msg3 = mi_archivo.readline()
print(msg3) 

mi_archivo.close()
```

```python
# Lectura de archivo, lecturas de todas las lineas
mi_archivo = open("mensajes.txt","r")

mensajes = mi_archivo.readlines()

mi_archivo.close()

for msg in mensajes:
  print(msg) # mostramos el mensaje
```
