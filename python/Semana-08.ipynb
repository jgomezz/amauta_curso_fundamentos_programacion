## 4.- Sentencia Elif (Sino Si)
Se encadena a un if u otro elif para comprobar múltiples condiciones, siempre que las anteriores no se ejecuten.

<center>
  <img width="500" src="https://raw.githubusercontent.com/jgomezm-tecsup/curso-introduccion-programacion/master/_assets/semana_08_if_elif.png" />
</center>


```python
#comando = "OTRA COSA"
comando = input("Ingresar el comando :")

if comando == "ENTRAR":
    print("Bienvenido al curso")
elif comando == "SALUDAR":
    print("Hola, espero que estes aprendiendo condicionales en Python")
elif comando == "SALIR":
    print("Saliendo del curso...")
else:
    print("Este comando no se reconoce")
```

```python
nota = int(input("Introduce una nota: "))
if nota >= 18:
    print("Sobresaliente")
elif nota >= 16:
    print("Notable")
elif nota >= 14:
    print("Bien")
elif nota >= 13:
    print("Suficiente")
else:
    print("Insuficiente")
```

Es posible simular el funcionamiento de elif con if utilizando expresiones condicionales

```python
nota = int(input("Introduce una nota: "))
if nota >= 18:
    print("Sobresaliente")
if  16 <= nota and nota < 18:
    print("Notable")
if nota >= 14 and nota < 16:
    print("Bien")
if nota >= 13 and nota < 14:
    print("Suficiente")
if nota < 13:
    print("Insuficiente")
```
