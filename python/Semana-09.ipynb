## 1.- Introducción

Las estructuras de control repetitivas permiten realizar una misma tarea varias veces hasta que una decisión deje de ser valida, cada vez que se repite una tarea, esta acción es conocio como iteración. Se trata de estructuras muy importantes, porque son las encargadas de controlar la ejecución repetitiva de un fragmento del código en un programa. 


## 2.- Sentencia **while** (Mientras)
Se basa en repetir un bloque a partir de evaluar una condición lógica, siempre que ésta sea verdadera (True).

Es importante indicar que la ejecución de cada iteracción debe afectar la evaluación lógica para determinar el fin de las iteracciones, en caso contrario se llegara a un bucle infinito.


```python
contador = 0
while contador <= 5:
    print("Contador = {}".format(contador))
    contador+=1
```

### 3.1.- Sentencia **else** en bucle **while**
Podemos ejecutar una bloque de código cuando el flujo repetitivo de While ha finalizado 


```python
contador = 0
while contador <= 5:
    print("Contador = {}".format(contador))
    contador+=1
else:
    print("Se ha completado las iteraciones y contador es {}".format(contador))
```

### 3.2.- Instrucción **break**
Sirve para "romper" la ejecución del While en cualquier momento. No se ejecutará el Else, ya que éste sólo se llama al finalizar la iteración.



```python
# Bucle While
contador = 0
while contador <= 5:
    if contador == 4:
        print("Rompemos el bucle cuando Contador es {}".format(contador))
        break
    print("Contador = {}".format(contador))
    contador +=1
```

```python
# Sentencia Else en bucle While
contador = 0
while contador <= 5:
    contador +=1
    if contador == 4:
        print("Rompemos el bucle cuando Contador es {}".format(contador))
        break
    print("Contador = {}".format(contador))
else:
    print("Se ha completado las iteraciones y contador es {}".format(contador))
```

### 3.3.- Instrucción **continue**
Sirve para "saltarse" la iteración actual sin romper el bucle.


```python
# Bucle While
contador = 0
while contador <= 5:
    contador +=1
    if contador == 3 or contador == 4:
        print("Continuamos con la siguiente iteración, NO SE procesa contador = {}".format(contador))
        continue
    print("Procesando contador = {} ....! ".format(contador))
```

```python
# Sentencia Else en bucle While
contador = 0
while contador <= 5:
    contador +=1
    if contador == 3 or contador == 4:
        print("Continuamos con la siguiente iteración, NO SE procesa contador = {}".format(contador))
        continue
    print("Procesando contador = {} ....! ".format(contador))
else:
    print("Se ha completado las iteraciones y contador es {}".format(contador))
```

### 3.4.- Sentencia **for** para diccionarios

```python
# Notas finales de un curso, solo recorre claves
notas_finales = {"MATE":15,"LENG":14, "HIST":18}

# Forma 1
print("Forma 1 :")
for clave in notas_finales:
  print(clave)

# Forma 2
print("Forma 2 :")
for clave in notas_finales.keys():
  print(clave)  
```

```python
# Notas finales de un curso, solo recorre valores
notas_finales = {"MATE":15,"LENG":14, "HIST":18}
for valor in notas_finales.values():
  print(valor)
```

```python
# Notas finales de un curso, recorre claves y valor
notas_finales = {"MATE":15,"LENG":14, "HIST":18}
for clave, valor in notas_finales.items():
  print(" {} --> {}".format(clave, valor))
```

## 4.- Sentencia **for** -  **range** 
Permiten ejecutar la iteración sobre un patron que se define con la fucion **range**


```python
# Imprimir los 5 primeros números naturales, empezando en 0
for nro in range(5): 
  print(nro)
```

```python
# Imprimir números de 10 a 15
for nro in range(10,16):
  print(nro)
```

```python
# Imprimir números de 10 a 15, incrementados de 2 en 2 
for nro in range(10,16,2):
  print(nro)
```

```python
# Ejercicio 16 : Imprimir los numeros de 2 a 10, y calcular su suma
suma = 0
for nro in range(2,11):
  suma = suma + nro
  print(nro)

print("La suma es {}".format(suma))
```

```python
# Ejercicio 17 : Imprimir los números de 2 a 10, incrementados en 3, y calcular 
#                la suma de esos números
suma = 0
for nro in range(2,11,3): # El tercer parámetro es el incremento
  suma = suma + nro
  print(nro)

print("La suma es {}".format(suma))
```
