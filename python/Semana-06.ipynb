# Introducción
El nuevo tipo de "estructura de datos" que veremos son los "set" que almacena datos desordenados. Entre sus caracteristicas es que no soporta duplicados en usa secuencia y permite realizar operaciones matemáticas como intersección, unión, diferencia y diferencia asimetrica.

Los Set se crean usando llaves de la siguiente manera:

> a = **{** $ a_{0},a_{1},a_{2},\dots $ **}**

A continuación mostraremos las diferentes propiedades de los "set"


```
numeros1 = [1,2,3,4]   # Lista --> Si se puede cambiar los valores
numeros2 = (1,2,3,4)   # Tupla --> No se puede cambiar los valores

# Definición de un set
numeros3 = {1,2,3,4}   # Set.  --> No permite datos repetidos y se pueden cambiar
print(numeros3)


estudiantes_c12 = {"Castro", "Jean Carlos", "Renzo"}
```

## add()

Agregar un dato al set

```
mis_datos = {10,20,30,40}
print(mis_datos)

mis_datos.add(50)
print(mis_datos)

mis_datos.add("Jaime Gomez")
print(mis_datos)

mis_datos.add((60,70))
print(mis_datos)

#mis_datos.add([80,90) # Porque es una lista, y no esta permitido en un set
#print(mis_datos)

#mis_datos.add({100,110}) # Porque es un set, y no esta permitido en un set
#print(mis_datos)
```

## update()

Permite agregar una tupla,lista o set a otro set

```
# Set unido con otro set
colores = {"red", "green", "yellow", "blue"}
nuevos_colores = {"white", "black", "purple"} # Es un set

#colores = colores + nuevos_colores # Esto da un error
colores.update(nuevos_colores)
print("colores :", colores)
```

```
# Set unido con una tupla
colores = {"red", "green", "yellow", "blue"}
nuevos_colores = ("white", "black", "purple") # Es una tupla

colores.update(nuevos_colores)
print("colores :", colores)
```

```
# Set unido con una list
colores = {"red", "green", "yellow", "blue"}
nuevos_colores = ["white", "black", "purple"] # Es una lista

colores.update(nuevos_colores)
print("colores :", colores)
```

## pop()

remueve un valor arbitrariamente del "set".

```
mis_datos = {10,20,30,40,60,70}
print(mis_datos)

valor = mis_datos.pop() # remueve un valor arbitrariamente
print(mis_datos)
print(valor)
```

## discard()

Remueve un dato por su valor.

```
mis_datos = {10,20,30,40}
print(mis_datos)

mis_datos.discard(10) # remueve el valor 10
print(mis_datos)

mis_datos.discard(10) # remueve el valor 10 por segunda vez
print(mis_datos)
```

## remove()

Remueve un dato por su valor, si no existe da una excepción.

```
mis_datos = {10,20,30,40}
print(mis_datos)

mis_datos.remove(20)
print(mis_datos)

#mis_datos.remove(20) # remueve por segunda vez el valor 20
#print(mis_datos)
```

## clear()

limpia un "set".

```
mis_datos = {10,20,30,40,60,70}
print(mis_datos)

mis_datos.clear() # remueve todos los valores
print(mis_datos)

#otros_datos = {}
#print(otros_datos)
```

## copy()

limpia un "set".

```
mis_datos = {10,20,30,40,60,70}
print("mis_datos =", mis_datos)

backup_datos = mis_datos.copy() # hace una copia de respaldo
print("backup_datos =", backup_datos)

print("Borra datos de la variable mis_datos")
mis_datos.clear()

print("mis_datos =", mis_datos)
print("backup_datos =", backup_datos)
```

## Operadores de pertenencia
- Para saber si un elemento pertenece a un conjunto se usa el operador : **in**
- Para saber si un elemento NO pertenece a un conjunto se usa el operador : **not in**


```
# Operador "in"
deportes = {"Futbol","Tenis","Baloncesto","Squash"}
print("Deportes :", deportes)

# Existe el deporta 'Disparo'?
existe_disparo = "Disparo" in deportes
print("Existe el deporte 'Disparo':", existe_disparo)

# Existe el deporta 'Futbol'?
existe_futbol = "Futbol" in deportes
print("Existe el deporte 'Futbol' :", existe_futbol)
```

```
# Operador "not in"
deportes = {"Futbol","Tenis","Baloncesto","Squash"}
print("Deportes :", deportes)

# No existe el deporta 'Disparo'?
no_existe_disparo = "Disparo" not in deportes
print("No existe el deporte 'Disparo':", no_existe_disparo)

# No existe el deporta 'Futbol'?
no_existe_futbol = "Futbol" not in deportes
print("No existe el deporte 'Futbol' :", no_existe_futbol)
```

## Unión
Permite unir 2 o mas conjuntos de datos ( $A \cup B$ ) y se denota por " **|** "

```
nros_multiplos_2 = { 2, 4, 6, 8}
nros_multiplos_4 = { 4, 8,12,16}

union_nros = nros_multiplos_2 | nros_multiplos_4 # Se parece al "update"
print(union_nros)
```

## Intersección
Permite ver los elementos comúnes entre 2 o mas conjuntos de datos ( $A \cap  B$ ) y se denota por " **&** "

```
nros_multiplos_2 = { 2, 4, 6, 8}
nros_multiplos_4 = { 4, 8,12,16}

interseccion_nros = nros_multiplos_2 & nros_multiplos_4 
print(interseccion_nros)
```

## Diferencia
Permite ver los elementos de un conjunto de datos  que no se encuentra en otro cojunto de datos ($ A - B$) . Se denota por " **-** ".

```
nros_multiplos_2 = { 2, 4, 6, 8}
nros_multiplos_4 = { 4, 8,12,16}

diferencia_nros = nros_multiplos_2 - nros_multiplos_4 
print(diferencia_nros)
```

## Diferencia Simétrica
Permite ver los elementos de un conjunto de datos  que no son comunes entre los conjunto de datos ($ A \vartriangle B$) . Se denota por " **^** ".

```
nros_multiplos_2 = { 2, 4, 6, 8}
nros_multiplos_4 = { 4, 8,12,16}

diferencia_asimetrica_nros = nros_multiplos_2 ^ nros_multiplos_4 
print(diferencia_asimetrica_nros)
```
